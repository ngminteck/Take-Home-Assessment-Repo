```python
import os
import sqlite3
import pandas as pd
your_sql_filepath = "Assignments/assets/task2.db"

try:
    conn = sqlite3.connect(your_sql_filepath)
except Exception as e:
    print(e)

# Get the table names
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
table_names = cursor.fetchall()
print(f"Table names: {table_names}")

# Read data from a specific table (replace 'your_table_name' with the actual table name)
tcdataitemtoenvtype = pd.read_sql_query("SELECT * FROM tcdataitemtoenvtype", conn)
tcenvtype = pd.read_sql_query("SELECT * FROM tcenvtype", conn)
tcdataitem = pd.read_sql_query("SELECT * FROM tcdataitem", conn)
tcsector = pd.read_sql_query("SELECT * FROM tcsector", conn)
tcperioddata = pd.read_sql_query("SELECT * FROM tcperioddata", conn)
tcperiodtextdata = pd.read_sql_query("SELECT * FROM tcperiodtextdata", conn)
tcperiod = pd.read_sql_query("SELECT * FROM tcperiod", conn)
tcsectorperioddata = pd.read_sql_query("SELECT * FROM tcsectorperioddata", conn)
tcsectorperiodtextdata = pd.read_sql_query("SELECT * FROM tcsectorperiodtextdata", conn)
tcsectorperiod = pd.read_sql_query("SELECT * FROM tcsectorperiod", conn)

tcdataitemtoenvtype.rename(columns={'index': 'ix_tcdataitemtoenvtype_index'}, inplace=True)
tcenvtype.rename(columns={'index': 'ix_tcenvtype_index'}, inplace=True)
tcdataitem.rename(columns={'index': 'ix_tcdataitem_index'}, inplace=True)
tcsector.rename(columns={'index': 'ix_tcsector_index'}, inplace=True)
tcperioddata.rename(columns={'index': 'ix_tcperioddata_index'}, inplace=True)
tcperiodtextdata.rename(columns={'index': 'ix_tcperiodtextdata_index'}, inplace=True)
tcperiod.rename(columns={'index': 'ix_tcperiod_index'}, inplace=True)
tcsectorperioddata.rename(columns={'index': 'ix_tcsectorperioddata_index'}, inplace=True)
tcsectorperiodtextdata.rename(columns={'index': 'ix_tcsectorperiodtextdata_index'}, inplace=True)
tcsectorperiod.rename(columns={'index': 'ix_tcsectorperiod_index'}, inplace=True)

s_df = tcsectorperioddata.merge(tcsectorperiodtextdata, on=['dataitemid','sectorperiodid','pacvertofeedpop'], how='outer')
s_df = s_df.merge(tcsectorperiod, on='sectorperiodid', how='outer')

c_df = tcperioddata.merge(tcperiodtextdata, on=['dataitemid','periodid','pacvertofeedpop'], how='outer')
c_df = c_df.merge(tcperiod, on='periodid', how='outer')

r_df = tcdataitemtoenvtype.merge(tcdataitem, on='dataitemid', how='outer')
r_df = r_df.merge(tcenvtype, on='tcenvtypeid', how='outer')

r_s_df = r_df.merge(s_df, on='dataitemid', how='outer')
r_s_df = r_s_df.merge(tcsector, on='tcsectorid', how='outer')

result_df = r_s_df.merge(c_df, on= ['dataitemid', 'dataitemvalue','dataitemvaluetext','pacvertofeedpop','institutionid', 'reportedcurrencyisocode','tcprimarysectorid','fiscalyear', 'periodenddate' ], how='outer')

print(result_df)
result_df.to_csv('Task2.csv', index=False)
print("Done saving to CSV.")

# Create a new sqlite connection to the output file
output_sql_filepath = "Task2.db"
output_conn = sqlite3.connect(output_sql_filepath)

# Write the DataFrame to the new sqlite file
result_df.to_sql('result_table', output_conn, if_exists='replace', index=False)

print("Done saving to SQL.")
# Close the connection
conn.close()
```
