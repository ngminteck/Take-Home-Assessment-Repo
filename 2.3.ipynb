```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming you have data on job relevance (closely related vs. unrelated)
# Modify the following code to match your actual data
relevant_jobs_x = 83 + 53 + 47 + 12
somewhat_relevant_jobs_x = 30 + 23 + 12 + 8
unrelated_jobs_x = 21
relevant_jobs_y = 23 + 9 + 32 + 15
somewhat_relevant_jobs_y = 3 + 7 + 22 + 11
unrelated_jobs_y = 28

labels = ['Closely Related', 'Somewhat Related', 'Unrelated']
sizes_x = [relevant_jobs_x, somewhat_relevant_jobs_x, unrelated_jobs_x]
sizes_y = [relevant_jobs_y, somewhat_relevant_jobs_y, unrelated_jobs_y]

plt.figure(figsize=(8, 6))
plt.pie(sizes_x, labels=labels, colors=['green', 'yellow', 'red'], autopct='%1.1f%%', startangle=90)
plt.title('Job Relevance: Group X & Group Y')
plt.axis('equal')
plt.show()
```

```python
# Assuming you have the total student count for each group
# Modify the following code to match your actual data
total_students_x = relevant_jobs_x + somewhat_relevant_jobs_x + unrelated_jobs_x
total_students_y = relevant_jobs_y + somewhat_relevant_jobs_y + unrelated_jobs_y

plt.figure(figsize=(6, 4))
plt.bar(['Group X', 'Group Y'], [total_students_x, total_students_y], color=['blue', 'orange'])
plt.xlabel('Student Group')
plt.ylabel('Total Students')
plt.title('Total Students: Group X vs. Group Y')
plt.show()
```
