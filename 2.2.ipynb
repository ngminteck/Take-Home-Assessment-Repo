```python
import geopandas as gpd

# Filepath to your KML file
filepath = "WirelessHotSpotsKML.kml"

# Read KML file into a GeoDataFrame
gdf = gpd.read_file(filepath)

# Print the first five rows
print(gdf.head())

# Plot the GeoDataFrame
gdf.plot()
```

```python
#Task 1: From the table, what are some of the information you can deduce for each hotspot?
#most hotspot in east area
```

```python
import pandas as pd

# Create a Pandas DataFrame from the GeoDataFrame
df = pd.DataFrame(gdf)
print(df.head())
```

```python
# Extract latitude and longitude from the geometry column
df['latitude'] = df.geometry.apply(lambda p: p.y)
df['longitude'] = df.geometry.apply(lambda p: p.x)
print(df.head())
```

```python
df.to_csv('wireless_hotspot.csv')
```
